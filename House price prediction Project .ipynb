# find outliers given dataset using Z score 

```python
# find limits


upper_limit = df[["sqft_above" , "sqft_basement" , "waterfront" , "view" , "condition" , "sqft_lot" , "bathrooms" , "bedrooms"]] .mean() + 3*df[["sqft_above" , "sqft_basement" , "waterfront" , "view" , "condition" , "sqft_lot" , "bathrooms" , "bedrooms"]].std()
lower_limit = df[["sqft_above" , "sqft_basement" , "waterfront" , "view" , "condition" , "sqft_lot" , "bathrooms" , "bedrooms"]] .mean() - 3*df[["sqft_above" , "sqft_basement" , "waterfront" , "view" , "condition" , "sqft_lot" , "bathrooms" , "bedrooms"]].std()

print("Upper Limit:  " , upper_limit)
print("Lower Limit:  " , lower_limit)
```

```python
from scipy.stats import zscore
```

```python
# Trim and find outliers 

columns_of_interest = ["sqft_above" , "sqft_basement" , "waterfront" , "view" , "condition" , "sqft_lot" , "bathrooms" , "bedrooms"]

z_scores = df[columns_of_interest].apply(zscore)

threshold = 3 
outlier_mask = (z_scores.abs() > threshold).any(axis=1)
clean_outliers = df[outlier_mask]

print("before removing outliers: ", len(df))
print("after removing outliers: ", len(clean_outliers))
print("outliers:" , len(df) - len(clean_outliers))
```

```python
clean_outliers
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_col = ["street" , "city" , "statezip" , "country"]
LE = LabelEncoder()
for col in categorical_col:
    clean_outliers[col] = LE.fit_transform(clean_outliers[col])
```

```python
clean_outliers.head()
```

```python
X = clean_outliers.iloc[: , 1:]
X
```

```python
y = clean_outliers["price"]
y
```

# Standarized the data

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
scale.fit_transform(X)
```

# Perform RFE for feature selection

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
rfe = RFE(estimator = LinearRegression() , n_features_to_select = 10)
X_rfe = rfe.fit_transform(X , y)
x = pd.DataFrame(X_rfe)
x
```

# split the data into train and test

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x ,y ,test_size =0.20 , random_state =0)
```

# perform model Linear regression

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
pred1 = LR.predict(x_test)
pred1
```

# Calculate accuracy by using MAE and r2_score

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score
```

```python
MAE = mae(y_test , pred1)
rmse = np.sqrt(MAE)
score = r2_score(y_test , pred1)



print("Mean Squared Error: ", MAE)
print("Root Mean Squared Error: ", rmse)
print("R2 score: ", score)
```

# 2nd model Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

RF = RandomForestRegressor()
RF.fit(x_train,y_train)
```

```python
pred2 = RF.predict(x_test)
pred2
```

```python
MAE1 = mae(y_test , pred2)
rmse1 = np.sqrt(MAE1)
score1 = r2_score(y_test , pred2)



print("Mean Squared Error: ", MAE1)
print("Root Mean Squared Error: ", rmse1)
print("R2 score: ", score1)
```
